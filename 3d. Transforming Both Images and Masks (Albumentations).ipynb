```python
#Import relevant libraries

import numpy as np
import time
import cv2
import os
import matplotlib.pyplot as plt
import albumentations as A
import random
```

```python
IMG_PATH=r'C:\Users\mpe_03\Desktop\masks\backpack_images'
MASK_PATH=r'C:\Users\mpe_03\Desktop\masks\backpack_masks'


def flip(IMG_PATH):
    
    #load the input image and associated mask
    for file in os.listdir(IMG_PATH):
        image_path=os.path.join(IMG_PATH, file)
        image_filename=file.split('.')[0]
        image=cv2.imread(image_path)
    
      
        mask_path=os.path.join(MASK_PATH, file)
        mask_filename=file.split('.')[0]
        mask=cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        
        
        aug=A.HorizontalFlip(p=1)
        
        augmented=aug(image=image, mask=mask)
        image_h_flipped=augmented['image']
        mask_h_flipped=augmented['mask']
        
        cv2.imwrite(f'C:/Users/mpe_03/Desktop/masks/img_transformations/{image_filename}_flip.jpg', image_h_flipped)
        cv2.imwrite(f'C:/Users/mpe_03/Desktop/masks/mask_transformations/{mask_filename}_flip.jpg', mask_h_flipped)

flip(IMG_PATH)


def elastic_transform(IMG_PATH):
    
    #load the input image and associated mask
    for file in os.listdir(IMG_PATH):
        image_path=os.path.join(IMG_PATH, file)
        image_filename=file.split('.')[0]
        image=cv2.imread(image_path)
    
      
        mask_path=os.path.join(MASK_PATH, file)
        mask_filename=file.split('.')[0]
        mask=cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        
        
        #aug=A.ElasticTransform(p=1, alpha=120,sigma=120*0.05, alpha_affine=120*0.03)
        aug=A.ElasticTransform(p=1, alpha=2000,sigma=1200*0.05, alpha_affine=1200*0.03)
        
        random.seed(42)
        augmented=aug(image=image, mask=mask)
        
        image_elastic=augmented['image']
        mask_elastic=augmented['mask']
        
        cv2.imwrite(f'C:/Users/mpe_03/Desktop/masks/img_transformations/{image_filename}_elastic.jpg', image_elastic)
        cv2.imwrite(f'C:/Users/mpe_03/Desktop/masks/mask_transformations/{mask_filename}_elastic.jpg', mask_elastic)

elastic_transform(IMG_PATH)

def mix(IMG_PATH):
    
    #load the input image and associated mask
    for file in os.listdir(IMG_PATH):
        image_path=os.path.join(IMG_PATH, file)
        image_filename=file.split('.')[0]
        image=cv2.imread(image_path)
    
      
        mask_path=os.path.join(MASK_PATH, file)
        mask_filename=file.split('.')[0]
        mask=cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        
        
        aug=A.Compose([
           A.HorizontalFlip(p=0.5),
           A.Rotate(limit=10, interpolation=1, border_mode=4, value=None, mask_value=None, always_apply=False, p=1),
            A.ElasticTransform(p=1, alpha=30,sigma=200*0.05, alpha_affine=200*0.03),
            A.GridDistortion(num_steps=5,border_mode=5,p=1)])
                
        random.seed(42)
        augmented=aug(image=image, mask=mask)
        
        image_mix=augmented['image']
        mask_mix=augmented['mask']
        
        cv2.imwrite(f'C:/Users/mpe_03/Desktop/masks/img_transformations/{image_filename}_mix1.jpg', image_mix)
        cv2.imwrite(f'C:/Users/mpe_03/Desktop/masks/mask_transformations/{mask_filename}_mix1.jpg', mask_mix)

mix(IMG_PATH)
      
```
