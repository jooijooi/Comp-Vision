```python
#importing of relevant Libraries
import numpy as np
import time
import cv2
import os
import matplotlib.pyplot as plt
```

```python
IMG_PATH=r'C:\Users\mpe_03\Desktop\masks\backpack_images'
MASK_PATH=r'C:\Users\mpe_03\Desktop\masks\backpack_masks'


def foreground(IMG_PATH):
    
    #load the input image and associated mask
    for file in os.listdir(IMG_PATH):
        image_path=os.path.join(IMG_PATH, file)
        image_filename=file.split('.')[0]
        image=cv2.imread(image_path)
      
      
        mask_path=os.path.join(MASK_PATH, file)
        mask_filename=file.split('.')[0]
        mask=cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
   
        

        #apply bitwise mask to show what the rough, approximate mask would give
        rough_output=cv2.bitwise_and(image,image, mask=mask)

        #any mask values greater than zero is set to probable foreground
        #values tweaked values based on mask pixel values
        mask[mask>250]=cv2.GC_PR_FGD
        mask[mask>40]=cv2.GC_BGD


        #allocate memory for the two arrays that the Grabcut algorithm internally uses when segmenting the foreground from the backgroun
        fgModel=np.zeros((1,65), dtype='float64')
        bgModel=np.zeros((1,65), dtype='float64')

        #apply Grabcut using mask segmentation method
        start=time.time()
    

        (mask, bgModel, fgModel)=cv2.grabCut(image, mask, None, bgModel,
                                             fgModel, iterCount=3, mode=cv2.GC_INIT_WITH_MASK)
        
        end=time.time()
        
        print('[INFO] applying Grabcut took{:.2f}seconds'.format(end-start))

        #the output mask has four possible output values
        values=(("Definite Background", cv2.GC_BGD),("Probable Background", cv2.GC_PR_BGD),("Definite Foreground", cv2.GC_FGD),("Probable Foreground", cv2. GC_PR_FGD),
        )

        #loop over possible Grabcut mask values
        for (name, value) in values:
            #construct a mask for the current value
            print("[INFO] showing mask for '{}'".format(name))
            valueMask=(mask==value).astype("uint8")*255

        #set all definite background and probable background pixels to 0
        #set all definite foreground and probable foreground pixels to 1
        #scale the mask from range[0,1] to[0,255]

        outputMask=np.where((mask==cv2.GC_BGD) | (mask==cv2.GC_PR_BGD), 0,1)
        outputMask=(outputMask*255).astype("uint8")

        #apply a bitwise AND to the image using mask generated by Grabcut to generate  final output image
        output=cv2.bitwise_and(image, image, mask=outputMask)

        cv2.imwrite(f'C:/Users/mpe_03/Desktop/masks/backpack_cutouts/{image_filename}.jpg', output)
        

foreground(IMG_PATH)

```
